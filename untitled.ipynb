```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

```python
import pandas as pd

url = "https://files.grouplens.org/datasets/movielens/ml-100k/u.data"

ratings = pd.read_csv(
    url,
    sep="\t",
    names=["user_id","movie_id","rating","timestamp"]
)

ratings.head()
```

```python
ratings = ratings.drop("timestamp", axis=1)

ratings.head()
```

```python
user_movie_matrix = ratings.pivot_table(
    index='user_id',
    columns='movie_id',
    values='rating'
)

user_movie_matrix.head()
```

```python
user_movie_matrix_filled = user_movie_matrix.fillna(0)
```

```python
user_similarity = cosine_similarity(user_movie_matrix_filled)

user_similarity_df = pd.DataFrame(
    user_similarity,
    index=user_movie_matrix.index,
    columns=user_movie_matrix.index
)

user_similarity_df.head()
```

```python
def recommend_movies(user_id, n_recommendations=5):

    # Find similar users
    similar_users = user_similarity_df[user_id].sort_values(
        ascending=False
    )[1:6]

    # Ratings from similar users
    similar_users_ratings = user_movie_matrix.loc[similar_users.index]

    # Average movie ratings
    movie_scores = similar_users_ratings.mean().sort_values(
        ascending=False
    )

    # Top recommendations
    return movie_scores.head(n_recommendations)
```

```python
recommend_movies(1)
```
