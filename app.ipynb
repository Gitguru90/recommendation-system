```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

movies = pd.DataFrame([
    {"movie_id": 1, "title": "The Matrix", "overview": "A hacker discovers reality is a simulation and joins a rebellion."},
    {"movie_id": 2, "title": "Inception", "overview": "A thief enters dreams to steal secrets and plant ideas."},
    {"movie_id": 3, "title": "Interstellar", "overview": "A team travels through a wormhole to save humanity."},
    {"movie_id": 4, "title": "The Dark Knight", "overview": "Batman fights the Joker in Gotham City."},
    {"movie_id": 5, "title": "John Wick", "overview": "An ex-hitman seeks revenge after a tragic loss."},
    {"movie_id": 6, "title": "Avatar", "overview": "A soldier on an alien world joins the natives."},
    {"movie_id": 7, "title": "Gladiator", "overview": "A betrayed general becomes a gladiator and seeks revenge."},
    {"movie_id": 8, "title": "Titanic", "overview": "A romance unfolds aboard the ill-fated ship Titanic."},
    {"movie_id": 9, "title": "Avengers: Endgame", "overview": "Heroes unite to reverse a universe-wide catastrophe."},
    {"movie_id": 10, "title": "The Lion King", "overview": "A young lion learns responsibility and reclaiming his kingdom."}
])

tfidf = TfidfVectorizer(stop_words="english")
tfidf_matrix = tfidf.fit_transform(movies["overview"])
similarity = cosine_similarity(tfidf_matrix)
movies.head()
```

```python
import os
import pickle

os.makedirs("model", exist_ok=True)

movies.to_csv("model/movies.csv", index=False)
with open("model/similarity.pkl", "wb") as f:
    pickle.dump(similarity, f)

print("Saved:", os.listdir("model"))
```

```python
app_code = r'''
import os
import pickle
import requests
import pandas as pd
import streamlit as st

st.set_page_config(page_title="Movie Recommender", layout="wide")

movie_csv = "model/movies.csv"
sim_path = "model/similarity.pkl"

if not os.path.exists(movie_csv) or not os.path.exists(sim_path):
    st.error("Missing files in model/. Run the data-creation cells first.")
    st.stop()

movies = pd.read_csv(movie_csv)
with open(sim_path, "rb") as f:
    similarity = pickle.load(f)

def fetch_poster(movie_id):
    return f"https://via.placeholder.com/500x750?text=Movie+ID+{movie_id}"

def recommend(movie):
    index = movies[movies["title"] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda x: x[1])

    recommended_movie_names = []
    recommended_movie_posters = []

    for i in distances[1:6]:
        movie_id = movies.iloc[i[0]].movie_id
        recommended_movie_names.append(movies.iloc[i[0]].title)
        recommended_movie_posters.append(fetch_poster(movie_id))

    return recommended_movie_names, recommended_movie_posters

st.title("Movie Recommender System")

movie_list = movies["title"].values
selected_movie = st.selectbox("Type or select a movie from the dropdown", movie_list)

if st.button("Show Recommendation"):
    names, posters = recommend(selected_movie)
    cols = st.columns(5)
    for i, col in enumerate(cols):
        with col:
            st.text(names[i])
            st.image(posters[i], use_container_width=True)
'''
with open("app.py", "w") as f:
    f.write(app_code)

print("app.py created")
```
